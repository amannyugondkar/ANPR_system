```python
!pip install easyocr
from ultralytics import YOLO
import cv2
import easyocr
import numpy as np 
reader = easyocr.Reader(['en'])    # Change/add language codes if needed
model = YOLO(r"C:\Users\cl502_19\Downloads\DL_ANPR\saved_models\finetune\indian_license_plate_best.pt")
img_path = r"indain_number_plate_testing/images/20230327_170242.jpg"

results = model.predict(img_path, conf=0.25, imgsz=640, save=False)
img = cv2.imread(img_path)

# pre-process
# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply sharpening
kernel = np.array([[0, -1, 0],
                   [-1, 5,-1],
                   [0, -1, 0]])
sharpened = cv2.filter2D(gray, -1, kernel)


# Optional: slight denoising
sharpened = cv2.fastNlMeansDenoising(sharpened, None, 10, 7, 21)


# Convert back to 3-channel for YOLO
img = cv2.cvtColor(sharpened, cv2.COLOR_GRAY2BGR)

for r in results:
    for i, box in enumerate(r.boxes.xyxy):
        x1, y1, x2, y2 = map(int, box)
        crop = img[y1:y2, x1:x2]
        ocr_result = reader.readtext(crop)
        text = ' '.join([res[1] for res in ocr_result])
        print(f"Detected Plate #{i+1}: {text}")
        cv2.imwrite(f"easyocr_crop_{i+1}.jpg", crop)

    def detect_plate(img_path):
        results = model.predict(img_path, conf=0.25, imgsz=640, save=False)
        img = cv2.imread(img_path)

    # Preprocess
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    kernel = np.array([[0, -1, 0],
                       [-1, 5, -1],
                       [0, -1, 0]])
    sharpened = cv2.filter2D(gray, -1, kernel)
    sharpened = cv2.fastNlMeansDenoising(sharpened, None, 10, 7, 21)
    img = cv2.cvtColor(sharpened, cv2.COLOR_GRAY2BGR)

    # OCR detection
    for r in results:
        for i, box in enumerate(r.boxes.xyxy):
            x1, y1, x2, y2 = map(int, box)
            crop = img[y1:y2, x1:x2]
            ocr_result = reader.readtext(crop)
            text = ' '.join([res[1] for res in ocr_result])
            print(f"Detected Plate #{i+1}: {text}")
            cv2.imwrite(f"easyocr_crop_{i+1}.jpg", crop)

# 🔹 Test any image path
detect_plate(r"C:\Users\cl502_19\Downloads\DL_ANPR\indain_number_plate_testing\images\20230327_170242.jpg")
detect_plate(r"C:\Users\cl502_19\Downloads\DL_ANPR\License-Plate-Recognition-11\test\images\00d9db3d2c186504_jpg.rf.5a493e083834aa4b4748f09a073cc200.jpg")
```
